# Cities For Doing Business

```python
leet_city_lines = lines.map(lambda x: (x[2].lower(),x[-9]))
```

```python
def proc_list(iterable):
    r = []
    for v1_iterable in iterable:
            r.append(v1_iterable)
    return tuple(r)
```

```python
def get_percentage(iterable):
    if len(iterable)>500:
        val = [k for k in iterable if k == 5]
        #percentage = float(len(val)/len(iterable))
        return (len(val)*100/len(iterable))
    else:
        return 0
```

```python
city_rating = leet_city_lines.groupByKey().mapValues(proc_list)
```

```python
best_cities_to_eat = city_rating.mapValues(get_percentage).sortBy(lambda x: x[1],ascending=False)
```

```python
cities_to_enjoy_food = best_cities_to_eat.collect()
```
